```python
from Bio.Blast import NCBIWWW
import random
import csv
```

```python
#Загружаем файл с последовательностью, которую хотим отобрать
file = open('PRNP_complete_CDS.txt', 'r')
prnp = file.readlines()
prnp
```

```python
#Разбиваем её на 22-мерные последовательности (потенциальные микроРНК) с шагом 1
micrna = []
for i in range(len(prnp[1])-22):
    micrna.append(prnp[1][i:i+22])
micrna
```

```python
len(micrna)
```

```python
#Для экономии времени случайным образом выбираем 40 последовательностей
#из нашего пула потенциальных микроРНК
#Для этого случайным образом сгенерируем 40 индексов (зафиксировав random.seed для воспроизводимости)
selected_ids = []
for i in range(40):
    random.seed(23+i) #Фиксируем random.seed для воспроизводимости
    num = random.randint(0,len(micrna))
    selected_ids.append(num)
selected_ids
```

```python
#Собираем в отдельный массив 40 последовательностей
selected_seqs = []
for i in selected_ids:
    selected_seqs.append(micrna[i])
selected_seqs
```

```python
#Бластим последовательности
filenames = []
for i in range(len(selected_seqs)):
    result_handle = NCBIWWW.qblast("blastn", "nt", selected_seqs[i])
    s = ['sequence', i+1, '.xml']
    filename = ''.join(str(x) for x in s)
    print(filename)
    filenames.append(filename)
#Записываем результат в файлы
    with open(filename, 'w') as save_file:
        blast_results = result_handle.read()
        save_file.write(blast_results)
```

```python
def twoone(list1):
    list2 = []
    for x in range(len(list1)):
        for y in range(len(list1[x])):
            list2.append(list1[x][y])
    return list2
#Отбираем последовательности, которые не выровнялись на другие последовательности в геноме человека
final_seqs = []
report = []
for i in range(len(selected_seqs)):
    s = ['sequence', i+1, '.xml']
    filename = ''.join(str(x) for x in s)
    file = open(filename, 'r')
    result = file.readlines()
    hs = []
    for string in result:
        if (string.find('Homo sapiens') != -1) and ((string.find('PRNP') == -1) \
        or (string.find('prion') == -1)):
            hs.append(string)
    file_list = []
    file_list.append(filename)
    report.append(file_list)
    report.append(hs)
    if hs == []:
        final_seqs.append(selected_seqs[i])
report = twoone(report)
#На всякий случай запишем найденные строки для каждой последовательности
#на случай, если захотим вручную посмотреть    
with open('final_report.csv', 'w') as file:
    for string in report:
        file.write(string+'\n')
#Запишем прошедшие фильтр последовательности микроРНК 
with open('final_micRNA.csv', 'w') as file:
    for string in final_seqs:
        file.write(string+'\n')
#Запишем выбранные изначально последовтаельности        
with open('selected_micRNA.csv', 'w') as file:
    for string in selected_seqs:
        file.write(string+'\n')
```
